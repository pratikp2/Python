```python
import numpy as np
import pandas as pd

# Libraries to help with data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Command to tell Python to actually display the graphs
%matplotlib inline
```

```python
data = pd.read_csv('Placement_Data.csv')
data.head()
```

```python
data['status'].value_counts()
```

```python
data.info()
```

```python
# hieghest number of missing value
data.isnull().sum()
```

```python
# etest
sns.histplot(data=data, x='etest_p')
```

```python
# hieghest correlation
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap='Spectral',vmin=-1,vmax=1);
```

```python
sns.histplot(data=data, x='ssc_p', y = 'hsc_p', hue='status');
```

```python
sns.lmplot(data=data,x='ssc_p',y='hsc_p',height=5,aspect=2,hue='status')
plt.show()
```

```python
sns.boxplot(data=data, x='ssc_p', y='status');
```
